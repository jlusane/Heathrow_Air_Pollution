```python
def bespoke_map_column_type(col):
    try:
        return float(col.strip().replace(',', '').replace('-', '0').replace('*', '0'))
    except ValueError:
        print('we need to remove this--->', col)
        return np.nan
    # capture case where pandas cannot perform string operations on the blank spaces
    except AttributeError:
        return np.nan
```

```python
# conver spells and expected to float for now... 
mortality_nhs['SPELLS'] = mortality_nhs['SPELLS'].str.replace('*', '0').astype('float')
mortality_nhs['OBSERVED'] = mortality_nhs['OBSERVED'].str.replace('*', '0').astype('float')
mortality_nhs['EXPECTED'] = mortality_nhs['EXPECTED'].str.replace('*', '0').astype('float')

admission_nhs['Commissioning Region/Provider'].fillna('Not Specified', inplace=True)
# admission_nhs['Diseases of the respiratory system'] = admission_nhs['Diseases of the respiratory system'].str.strip().str.replace(',', '').str.replace('-', '0').str.replace('*', '0')
admission_nhs['Diseases of the respiratory system'] = admission_nhs['Diseases of the respiratory system'].map(bespoke_map_column_type)
```

```python
def bespoke_method_2(arggg):
    if 'hillingdon' in arggg.strip().lower():
        return 'Hillingdon Trust'
    elif 'ealing' in arggg.strip().lower():
        return 'Ealing NHS Trust'
    elif 'ashford' in arggg.strip().lower():
        return 'Ashford NHS Trust'
    elif 'west middlesex' in arggg.strip().lower():
        return 'West Middlesex NHS Trust'
    elif 'frimley' in arggg.strip().lower():
        return 'Frimley NHS Trust'
    elif 'berkshire' in arggg.strip().lower():
        return 'Berkshire NHS Trust'
    elif 'hounslow' in arggg.strip().lower():
        return 'Hounslowe NHS Trust'
    elif 'teddington' in arggg.strip().lower():
        return 'Teddington NHS Trust'
    elif 'richmond' in arggg.strip().lower():
        return 'Richmond NHS Trust'
    elif 'clayponds' in arggg.strip().lower():
        return 'Clayponds NHS Trust'
    elif 'cygnet' in arggg.strip().lower():
        return 'Cygnet NHS Trust'
    elif 'cassel' in arggg.strip().lower():
        return 'Cassel NHS Trust'
    elif 'london north west university healthcare' in arggg.strip().lower():
        return 'London North West University Healthcare' 
    elif 'north west london hospitals nhs trust' in arggg.strip().lower():
        return 'North West London Hospitals NHS Trust'
    else:
        return 'Trust Not Specified'
    
    
    
```
